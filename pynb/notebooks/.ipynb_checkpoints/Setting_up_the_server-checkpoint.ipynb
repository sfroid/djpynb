<h1>Fixing notebook names</h1><hr />

If you have used nbconvert or nbviewer to convert your ipython notebook files to html format, you might have noticed that the title for the html file is missing.

Digging in, I found that the html_full template `html_full.tpl` has the field in place and is supposed to be filled from the `nb.metadata.name` field. However, all my notebooks have an empty string in this field.

What can we do now? Start debugging!!! I chased down the issue to the file `filenbmanager.py`, line 185, in the folder `lib/python2.7/site-packages/IPython/html/services/notebooks/`

```
    # Right before we save the notebook, we write an empty string as the
    # notebook name in the metadata. This is to prepare for removing
    # this attribute entirely post 1.0. The web app still uses the metadata
    # name for now.
    nb.metadata.name = u''
```

We are explicitly setting the name field to an *empty string*.

The fix is to simply comment out the line and try things out (but not before backing up your existing notebooks).

Everything seems to work fine after I finished this surgery... however, being no expert in ipython notebook internals, I am not sure if everything will continue to be fine. If someone more knowledgeable come across this and feels its the wrong way... do drop me a line at < ipythonIssues at csproto dot net >.

```
upstream sfroid_server {
  # fail_timeout=0 means we always retry an upstream even if it failed
  # to return a good HTTP response (in case the Unicorn master nukes a
  # single worker for timing out).
 
  server unix:/home/sfroid/website/django/run/gunicorn.sock fail_timeout=0;
}
 
server {
 
    listen   80;
    server_name sfroid.net;
 
    client_max_body_size 4G;
 
    access_log /home/sfroid/website/django/logs/nginx-access.log;
    error_log /home/sfroid/website/django/logs/nginx-error.log;
 
    location /static/ {
        alias   /home/sfroid/website/django/sfroid/static/;
    }
    
    location /media/ {
        alias   /home/sfroid/website/django/sfroid/media/;
    }
 
    location / {
        # an HTTP header important enough to have its own Wikipedia entry:
        #   http://en.wikipedia.org/wiki/X-Forwarded-For
        proxy_set_header X-Forwarded-For $proxy_add_x_forwarded_for;
 
        # enable this if and only if you use HTTPS, this helps Rack
        # set the proper protocol for doing redirects:
        # proxy_set_header X-Forwarded-Proto https;
 
        # pass the Host: header from the client right along so redirects
        # can be set properly within the Rack application
        proxy_set_header Host $http_host;
 
        # we don't want nginx trying to do something clever with
        # redirects, we set the Host: header above already.
        proxy_redirect off;
 
        # set "proxy_buffering off" *only* for Rainbows! when doing
        # Comet/long-poll stuff.  It's also safe to set if you're
        # using only serving fast clients with Unicorn + nginx.
        # Otherwise you _want_ nginx to buffer responses to slow
        # clients, really.
        # proxy_buffering off;
 
        # Try to serve static files from nginx, no point in making an
        # *application* server like Unicorn/Rainbows! serve static files.
        if (!-f $request_filename) {
            proxy_pass http://sfroid_server;
            break;
        }
    }
 
    # Error pages
    error_page 500 502 503 504 /500.html;
    location = /500.html {
        root /home/nagarajan/exps/djsvr/sfroid/static/;
    }
}
```


<h3>Special path for ipython notebook viewer</h3>

In the ipython notebook folder, create a subfolder called `html`. The ipython notebook files will be automatically converted to html files by our nbwatcher program (a python script we wrote) and the html file will be placed in the html folder.

We will now add a dedicated path for `pynb` in nginx pointing to this folder.

Open `/etc/nginx/sites-available/sfroidsvr` and add the following

```
    location /pynb {
        alias   /home/sfroid/website/pynb/html/;
        autoindex on;
    }
```


<h3>Enable the site sfroidsvr</h3>

Create link in the `sites-enabled` folder and restart nginx

```
sudo ln -s /etc/nginx/sites-available/sfroidsvr /etc/nginx/sites-enabled/sfroidsvr


# You might need to delete the `default`
# configuration from `sites-enabled` to make the djsvr site visible.
sudo rm /etc/nginx/sites-enabled/default

sudo service nginx restart
```

<h2>South crash course</h2>
South is used for migrating applications. It's not project based... it works on apps.

First, make sure we have run `syncdb` after we installed south.

To start off, we create the initial migration for an app
```
./manage.py schemamigration appname --initial
./manage.py migrate appname
```

Whenever we change the data models in `appname`, we need to create new migrations and apply them...

```
./manage.py schemamigration appname --auto
./manage.py migrate appname
```

Thats all there is to it.
