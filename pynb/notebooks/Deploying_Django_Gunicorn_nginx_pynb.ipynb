```
$ sudo apt-get install supervisor
```

Now, we add a configuration file for our django server app by adding a file `/etc/supervisor/conf.d/djsvr.conf` with the following data...

```
[program:djsvr]
command = /home/nagarajan/exps/djsvr/bin/gunicorn_start.sh
user = djsvruser
stdout_logfile = /home/nagarajan/exps/djsvr/logs/gunicorn_supervisor.log
redirect_stderr = true
```

We now add the log directory and log file

```
$ mkdir -p /home/nagarajan/exps/djsvr/logs
$ touch /home/nagarajan/exps/djsvr/logs/gunicorn_supervisor.log
```

```
upstream djsvr_server {
  # fail_timeout=0 means we always retry an upstream even if it failed
  # to return a good HTTP response (in case the Unicorn master nukes a
  # single worker for timing out).
 
  server unix:/home/nagarajan/exps/djsvr/run/gunicorn.sock fail_timeout=0;
}
 
server {
 
    listen   80;
    server_name mywebsite.com;
 
    client_max_body_size 4G;
 
    access_log /home/nagarajan/exps/djsvr/logs/nginx-access.log;
    error_log /home/nagarajan/exps/djsvr/logs/nginx-error.log;
 
    location /static/ {
        alias   /home/nagarajan/exps/djsvr/static/;
    }
    
    location /media/ {
        alias   /home/nagarajan/exps/djsvr/media/;
    }
 
    location / {
        # an HTTP header important enough to have its own Wikipedia entry:
        #   http://en.wikipedia.org/wiki/X-Forwarded-For
        proxy_set_header X-Forwarded-For $proxy_add_x_forwarded_for;
 
        # enable this if and only if you use HTTPS, this helps Rack
        # set the proper protocol for doing redirects:
        # proxy_set_header X-Forwarded-Proto https;
 
        # pass the Host: header from the client right along so redirects
        # can be set properly within the Rack application
        proxy_set_header Host $http_host;
 
        # we don't want nginx trying to do something clever with
        # redirects, we set the Host: header above already.
        proxy_redirect off;
 
        # set "proxy_buffering off" *only* for Rainbows! when doing
        # Comet/long-poll stuff.  It's also safe to set if you're
        # using only serving fast clients with Unicorn + nginx.
        # Otherwise you _want_ nginx to buffer responses to slow
        # clients, really.
        # proxy_buffering off;
 
        # Try to serve static files from nginx, no point in making an
        # *application* server like Unicorn/Rainbows! serve static files.
        if (!-f $request_filename) {
            proxy_pass http://djsvr_server;
            break;
        }
    }
 
    # Error pages
    error_page 500 502 503 504 /500.html;
    location = /500.html {
        root /home/nagarajan/exps/djsvr/static/;
    }
}
```


Create link in the `sites-enabled` folder and restart nginx

```
$ sudo ln -s /etc/nginx/sites-available/djsvr /etc/nginx/sites-enabled/djsvr

$ sudo service nginx restart
```

You might need to delete the `default` configuration from `sites-enabled` to make the djsvr site visible.


<h2>Lets add a special path for ipython notebook html files</h2>

In the ipython notebook folder, create a subfolder called `html`. The ipython notebook files will be automatically converted to html files by our nbwatcher program (a python script we wrote) and the html file will be placed in the html folder.

We will now add a dedicated path for `ipynb` in nginx pointing to this folder.

Open `/etc/nginx/sites-available/djsvr` and add the following

```
location /ipynb/ {
    alias   /home/nagarajan/exps/pynb/html/;
    autoindex on;
}
```

and then we restart nginx

```
sudo service nginx restart
```

<h2>Undoing all the hard work</h2>

Removing djsvr from nginx

```
$ sudo rm /etc/nginx/sites-enabled/djsvr
$ sudo service nginx restart
```

Remove djsvr from supervisor

```
$ sudo supervisorctl stop djsvr

(optional)
$ sudo rm /etc/supervisor/conf.d/djsvr.conf
```



<h3>the end...</h3>
