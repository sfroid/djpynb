<h1>Falling into the Sun</h1><hr />

<h3>How strongly does the sun pull us?</h3>
How long will will it take to fall into the sun from a distace equal to the earth's orbit?
Lets find out by writing a python program...


```python
# script to calculate time to reach sun...
from math import sqrt
from time import time

# physical constants/values
G = 6.67384e-11 # gravitational constant
MS = 1.98855e30 # mass of sun
esd = 149597870700.0 # earth-sun distance in meters

# program variables
n = 1000000
myconst1 = (G*MS*n*n)/(esd*esd)
myconst2 = esd/n

def getA(k):
    """ acceleration at distance (esd*k/n) from sun"""
    ak = myconst1/pow(n-k,2)
    return ak

def getV(v, a, t):
    """ velocity at distance (esd*k/n) from sun"""
    vk = v + a*t
    return vk

def getVAT(k, vold, aold, told):
    """ time to get from position (esd*k/n) to (esd*(k+1)/n)"""
    vk = getV(vold, aold, told)
    ak = getA(k)

    tk = (sqrt(vk*vk + 2*ak*myconst2) - vk)/ak
    return vk, ak, tk

stime = time()
sts, v, a, t = 0, 0, 0, 0
for x in range(n):
    v, a, t = getVAT(x, v, a, t)
    sts += t

print "Travel time = %.2f seconds\nDays = %5.3f"%(sts, sts/(60.0*60*24))
print "Computation time = %5.4f seconds"%(time() - stime)
```

<h2>Analytic calculation... </h2>
Can we calculate the same result by using Newton's laws and without using computers?

All we have to do is to solve the differential equation. So, lets jump to it.




<h3>The time taken by an object to fall into the sun is given by</h3>

$T = \pi \frac{d^{\frac{3}{2}}}{\sqrt{8 GM_s}} = 64.5629\text{ days}$.

<b>How in the hell did $\pi$ make its way into the answer??</b>

Let's back up a bit and derive the answer and maybe we will see what's going on... (warning/welcome: Maths ahead :-) )

Let us assume that the earth and the sun lie on the x axis and d is the distance between them. Also, at time t=0, let the earth be located at at x=0 and the sun at x=d.

We know from Newton's Law of Gravitation that the force on earth due to the sun is given by

$F = \frac{GM_sm_e}{r^2}$

where
$F$: force in newtons;<br />
$G$: gravitational constant ($6.67384 \times 10^{-11}\text{ kg}^{-1}\text{m}^3\text{s}^{-2}$);<br />
$M_s$: mass of the sun ($1.989 \times 10^{30}\text{ kg}$ );<br />
$m_e$: mass of the earth ($5.972 \times 10^{24}\text{ kg}$ );<br />
$r$: distance between earth and sun in meters.<br />
At $t=0, r = d = 1.496 \times 10^{11}\text{ m}$.<br />

Let's solve this equation to find the time taken to fall into the sun:

$F = m_ea = \frac{GM_sm_e}{(d-x)^2}$

$\implies a = \frac{GM_s}{(d-x)^2}$.

Now, we also know that

$a = \frac{dv}{dt} = \frac{d^2x}{dt^2}$

and we can also write 

$\frac{dv}{dt} = \frac{dv}{dx} \times \frac{dx}{dt} = \frac{dv}{dx} \times v$


We did the last step so that we can separate the variables and solve the differential equation. Now

$v \frac{dv}{dx} = \frac{GM_s}{(d-x)^2}$.

Moving dx to the right side and integrating, we get

$\int v\, dv = \int \frac{GM_s}{(d-x)^2} dx$.

Performing the integration, we get

$\frac{v^2}{2} = \frac{GM_s}{(d-x)} + C$

where C is the constant of integration. We find C by using the known condition that v = 0 at x = 0, which

$ \implies C = \frac{-GM_s}{d}$.

Putting this value of C in the equation for v,

$\frac{v^2}{2} = \frac{GM_s}{(d-x)} - \frac{GM_s}{d}$;

$v = \sqrt{2GM_s \left( \frac{1}{(d-x)} - \frac{1}{d} \right) }$;

$\frac{dx}{dt} = \sqrt{\frac{2GM_s}{d} . \frac{x}{(d-x)}}$.

Moving dt to the right and variables involving x to the left and integrating, we get

$\int_0^d \sqrt{\frac{d- x}{x}} dx = \int_0^T \sqrt{\frac{2GM_s}{d}} dt $.

I used Wolfram|Alpha to find the indefinite integral of the left side to get the following:

$\left. \sqrt{x(d-x)} - \frac{d}{2} tan^{-1} \left( \frac{(d-2x)} {2 \sqrt{x(d - x)}} \right) \right\vert_0^d $$= \sqrt{\frac{2GM_s}{d}} T$.


At both $x=0$ and $x=d$, the term inside the $tan^{-1}()$ parentheses contains a division by zero. That is not too bad as $tan^{-1}(\pm \infty)$ is defined. And to know whether the division by zero leads to a $+\infty$ or a $-\infty$, we need to evaluate the LHS at $x=0^+$ and $x=d^-$:

$\left. \sqrt{x(d-x)} - \frac{d}{2} tan^{-1} \left( \frac{(d-2x)} {2 \sqrt{x(d - x)}} \right) \right\vert_{0^+}^{d^-}$.

The term $\sqrt{x(d-x)}$ equals $0^+$at both $x=0^+$ and $x=d^-$,

$\therefore LHS = -\frac{d}{2} tan^{-1}(-\infty) + \frac{d}{2} tan^{-1}(+\infty) $<br />
$= -\frac{d}{2}(-\frac{\pi}{2}) + \frac{d}{2}(+\frac{\pi}{2})$<br />
$= \pi \frac{d}{2}$.

Putting the LHS and RHS together, we get

$\pi \frac{d}{2} = \sqrt{\frac{2GM_s}{d}} T$.

Rearranging the variables around, we get

$T = \pi \frac{d^{\frac{3}{2}}}{\sqrt{8GM_s}}$.

Putting in the values, we get

$T = 5578237\text{ seconds} = 64.5629\text{ days}$.

The computed answer is off by about 500 seconds from the analytical answer, but thats an error of less than 0.01%.

<h2>Lets plot the progress of the projectile on a graph...</h2>

Now, we want to plot the distance from sun vs time.

However, the equation that defines the relation between the distance and time cannot be easily solved to give the distance at a particular time. In that case, what can we do?

We use the equation for time...

$ T = \sqrt{\frac{d}{2GM_s}} \left( \sqrt{y(d-y)} - \frac{d}{2} tan^{-1} \left( \frac{(d-2y)} {2 \sqrt{y(d - y)}} \right) + \frac{d\pi}{4} \right)$.

and do binary search to find the y that gives the desired t.

For example, in order to find the position y at time t=10 seconds, we guess an initial value of y, say 500, and find the time corresponding to it.

If the time is larger than 10 seconds, we know that y must be smaller, so that it can be covered in 10 seconds. We decrease y in this case (increase y if t < 10). We perform the increases and decreases in progressively smaller quantities until we converge on the corect y.

```python
from math import sqrt, atan, pi

# physical constants/values
G = 6.67384e-11 # gravitational constant
MS = 1.989e30 # mass of sun
d = 149600000000.0 # earth-sun distance in meters

def getT(y):
    return sqrt(d/(2*G*MS)) * (sqrt(y*(d - y)) - (d/2.0)*atan((d - 2*y)/(2*sqrt(y*(d-y)))) + d*pi/4)
    

def findY(t):
    # given a particular t, find the value of y that fits
    a, b = 0.0, d
    error = 100
    steps = 0
    while steps < 200 and abs(error) > 0.1:
        ny = (a + b)/2.0
        nt = getT(ny)
        error = nt - t
        steps += 1
        
        if error > 0:
            b = ny
        else:
            a = ny
        
    if steps < 200:
        return ny
    return 0
        
from pylab import *

t = [z+0.01 for z in range(0, 55782, 10)]
y = [d - findY(z*100) for z in t]

figure()
plot(t, y, 'r')
xlabel('t')
ylabel('x')
title('path to sun')
show()

```
